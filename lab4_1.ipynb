```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import numpy as np
import timeit
import warnings
warnings.filterwarnings('ignore')
print("Setup Complete") # Setting up the coding environment
```

<p><h4>Виконую завдання першого рівня використовуючи pandas.DataFrame</h4></p>

```python
# Set the filepath to the dataset
filepath = 'data/household_power_consumption.txt'

# Read the dataset into dataframe
df = pd.read_csv(filepath, sep=';', header=0)

# Print the data types of each column
print(df.dtypes)

# Print the number of rows
print("The DataFrame has", df.shape[0], "rows")

# Display the first five rows
df.head()
```

```python
# A little bit of Data Cleaning
df.drop(df[df.eq('?').any(axis=1)].index, inplace=True)

# Changing type or columns
numeric_columns = ['Global_active_power', 'Global_reactive_power', 'Voltage', 'Global_intensity', 'Sub_metering_1', 'Sub_metering_2', 'Sub_metering_3']
df[numeric_columns] = df[numeric_columns].astype(float)
df['Date'] = pd.to_datetime(df['Date'], format='%d/%m/%Y')
df['Time'] = pd.to_datetime(df['Time'], format='%H:%M:%S').dt.time

# Print the data types of each column
print(df.dtypes)

# Print the number of rows
print("The DataFrame has", df.shape[0], "rows")

# Display the first five rows
df.head()
```

<p><h4>1. Обрати всі домогосподарства, у яких загальна активна споживана потужність перевищує 5 кВт</h4></p>

```python
start_time = timeit.default_timer()
households_over_5kw = df[df['Global_active_power'] > 5]
end_time = timeit.default_timer()
print(f"Time: {end_time - start_time} seconds")

# Print the number of rows
print("The DataFrame has", households_over_5kw.shape[0], "rows")

# Display the first five rows
households_over_5kw.head()
```

<p><h4>2. Обрати всі домогосподарства, у яких вольтаж перевищую 235 В</h4></p>

```python
start_time = timeit.default_timer()
voltage_over_235v = df[df['Voltage'] > 235]
end_time = timeit.default_timer()
print(f"Time: {end_time - start_time} seconds")

# Print the number of rows
print("The DataFrame has", voltage_over_235v.shape[0], "rows")

# Display the first five rows
voltage_over_235v.head()
```

<p><h4>3. Обрати всі домогосподарства, у яких сила струму лежить в межах  19-20 А, для них виявити ті, у яких пральна машина та холодильни  
споживають більше, ніж бойлер та кон.</h4></p>

```python
start_time = timeit.default_timer()
selected_households = df[(df['Global_intensity'] >= 19) & (df['Global_intensity'] <= 20)]
selected_households = selected_households[selected_households['Sub_metering_2'] > selected_households['Sub_metering_3']]
end_time = timeit.default_timer()
print(f"Time: {end_time - start_time} seconds")

# Print the number of rows
print("The DataFrame has", selected_households.shape[0], "rows")

# Display the first five rows
selected_households.head()
```

<p><h4>4. Обрати випадковим чином 500000 домогосподарств (без повторів елементів вибірки), для них обчислити середні величини усіх 3-х груп споживання електричної енергіїї</h4></p>

```python
start_time = timeit.default_timer()

# Selecting a random sample of 500,000 rows from the start DataFrame
random_households = df.sample(n=500000, replace=True)

# Calculation of average values
average_sub_metering_1 = random_households['Sub_metering_1'].mean()
average_sub_metering_2 = random_households['Sub_metering_2'].mean()
average_sub_metering_3 = random_households['Sub_metering_3'].mean()

print("Average energy consumption according to my calculations:")
print("Average for sub_metering_1:", average_sub_metering_1)
print("Average for sub_metering_2:", average_sub_metering_2)
print("Average for sub_metering_3:", average_sub_metering_3)

end_time = timeit.default_timer()
print(f"Time: {end_time - start_time} seconds")
```

<p><h4>5. Обрати ті домогосподарства, які після 18-00 споживають понад 6
кВт за хвилину в середньому, серед відібраних визначити ті, у яких
основне споживання електроенергії у вказаний проміжок часу
припадає на пральну машину, сушарку, холодильник та освітлення
(група 2 є найбільшою), а потім обрати кожен третій результат із
першої половини та кожен четвертий результат із другоої полвини.</h4></p>

```python
start_time = timeit.default_timer()

# Selection of households that consume more than 6 kW after 18:00:00
filtered_df = df[(df['Time'] > pd.to_datetime('18:00:00').time()) & (df['Global_active_power'] > 6)]

# Another filtering
filtered_df = filtered_df[(filtered_df['Sub_metering_2'] > filtered_df['Sub_metering_1']) & (filtered_df['Sub_metering_2'] > filtered_df['Sub_metering_3'])]

# Choose every third result from the first half and every fourth result from the second half
first_half = filtered_df.iloc[:len(filtered_df) // 2].iloc[2::3]
second_half = filtered_df.iloc[len(filtered_df) // 2:].iloc[3::4]

# Сoncatenation of parts
res = pd.concat([first_half, second_half])

# Print the number of rows
print("The DataFrame has", res.shape[0], "rows")

end_time = timeit.default_timer()
print(f"Time: {end_time - start_time} seconds")

# Display the first ten rows
res.head(10)
```

<p><h4>Виконую завдання першого рівня використовуючи NumPy.Array</h4></p>

```python
# Creating a DataFrame funtion
def createdf():
    filepath = 'data/household_power_consumption.txt'
    df = pd.read_csv(filepath,sep=';',header=0)
    df.drop(df[df.eq('?').any(axis=1)].index, inplace=True)

    numeric_columns = ['Global_active_power', 'Global_reactive_power', 'Voltage', 'Global_intensity', 'Sub_metering_1', 'Sub_metering_2', 'Sub_metering_3']
    df[numeric_columns] = df[numeric_columns].astype(float)
    df['Date'] = pd.to_datetime(df['Date'], format='%d/%m/%Y').dt.strftime('%Y-%m-%d')
    
    return df

# Calling the function and changing to NumPy Array
df = createdf()
datanp = df.to_numpy()
print(datanp)

print("The NumPy Array has", datanp.shape[0], "rows")
```

<p><h4>1. Обрати всі домогосподарства, у яких загальна активна споживана потужність перевищує 5 кВт</h4></p>

```python
start_time = timeit.default_timer()
households_over_5kw = datanp[datanp[:, 2] > 5]
end_time = timeit.default_timer()
print(f"Time: {end_time - start_time} seconds")

# Print the number of rows
print("The NumPy Array has", households_over_5kw.shape[0], "rows")

print(households_over_5kw)
```

<p><h4>2. Обрати всі домогосподарства, у яких вольтаж перевищую 235 В</h4></p>

```python
start_time = timeit.default_timer()
voltage_over_235v = datanp[datanp[:, 4] > 235]
end_time = timeit.default_timer()
print(f"Time: {end_time - start_time} seconds")

# Print the number of rows
print("The NumPy Array has", voltage_over_235v.shape[0], "rows")

print(voltage_over_235v[:5])
```

<p><h4>3. Обрати всі домогосподарства, у яких сила струму лежить в межах  19-20 А, для них виявити ті, у яких пральна машина та холодильни  
споживають більше, ніж бойлер та кон.</h4></p>

```python
start_time = timeit.default_timer()
selected_households = datanp[(datanp[:, 5] >= 19) & (datanp[:, 5] <= 20)]
selected_households = selected_households[selected_households[:, 7] > selected_households[:, 8]]
end_time = timeit.default_timer()
print(f"Time: {end_time - start_time} seconds")

# Print the number of rows
print("The NumPy Array has", selected_households.shape[0], "rows")

print(selected_households[:5])
```

<p><h4>4. Обрати випадковим чином 500000 домогосподарств (без повторів елементів вибірки), для них обчислити середні величини усіх 3-х груп споживання електричної енергіїї</h4></p>

```python
start_time = timeit.default_timer()

# Obtaining 500000 random indexes without repetitions
random_indices = np.random.choice(datanp.shape[0], size=500000, replace=True)

# selecting random rows from datanp using random indexes
random_households = datanp[random_indices]

# Obtaining values for each group
sub_metering_1_values = random_households[:, 6]  
sub_metering_2_values = random_households[:, 7]  
sub_metering_3_values = random_households[:, 8]  

# # Calculation of average values
average_sub_metering_1 = np.mean(sub_metering_1_values)
average_sub_metering_2 = np.mean(sub_metering_2_values)
average_sub_metering_3 = np.mean(sub_metering_3_values)

print("Average energy consumption according to my calculations:")
print("Average for sub_metering_1:", average_sub_metering_1)
print("Average for sub_metering_2:", average_sub_metering_2)
print("Average for sub_metering_3:", average_sub_metering_3)

end_time = timeit.default_timer()
print(f"Time: {end_time - start_time} seconds")
```

<p><h4>5. Обрати ті домогосподарства, які після 18-00 споживають понад 6
кВт за хвилину в середньому, серед відібраних визначити ті, у яких
основне споживання електроенергії у вказаний проміжок часу
припадає на пральну машину, сушарку, холодильник та освітлення
(група 2 є найбільшою), а потім обрати кожен третій результат із
першої половини та кожен четвертий результат із другоої полвини.</h4></p>

```python
start_time = timeit.default_timer()

# Filtering
filtered_data_np = datanp[(datanp[:, 1] > '18:00:00') & (datanp[:, 2] > 6) & (datanp[:, 7] > datanp[:, 6]) & (datanp[:, 7] > datanp[:, 8])]

# Dividing data into 2 parts
group1 = filtered_data_np[:len(filtered_data_np)//2]
group2 = filtered_data_np[len(filtered_data_np)//2:]

# Selecting every third result from the first half
every_third_group1 = group1[::3]

# Selecting every fourth result from the second half
every_fourth_group2 = group2[::4]

# Concatenation of 2 parts
final_results = np.concatenate((every_third_group1, every_fourth_group2), axis=0)


end_time = timeit.default_timer()
print(f"Time: {end_time - start_time} seconds")

# Print the number of rows
print("The NumPy Array has", final_results.shape[0], "rows")

print(final_results[:5])
```
